La central de un servicio de emergencia dispone de 3 telefonos para atender las llamadas. La duracion de las llamadas sigue una distribucion normal con los parametros que recoge la tabla en minutos.

| Operador telefono | media | desviacion |
|-----------------------|--------|--------------|
|               1                        | 2.0       |    0.4              |
|               2                        | 2.1       |    0.3               |
|               3                        | 1.9       |    0.5              |

Determine el tiempo medio de respuesta de la central de emergencias

```python
import numpy as np
import matplotlib.pyplot as plt

media = [2.0, 2.1, 1.9]
desviacion = [0.4, 0.3, 0.5]

L = []
for _ in range(100000):
    (x, y, z) = np.random.normal(media, desviacion)
    L.append(z)

plt.hist(L, bins=500, edgecolor='blue')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de L')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal para cada operador
medias = [2.0, 2.1, 1.9]
desviaciones = [0.4, 0.3, 0.5]

num_simulaciones = 100
num_llamadas = 100

tiempos_respuesta_central = []

for _ in range(num_simulaciones):
    tiempos_respuesta_operadores = []

    for _ in range(num_llamadas):
        tiempos_llamadas = np.random.normal(medias, desviaciones)
        tiempo_respuesta_operador = np.min(tiempos_llamadas)
        tiempos_respuesta_operadores.append(tiempo_respuesta_operador)

    tiempo_respuesta_central = np.mean(tiempos_respuesta_operadores)
    tiempos_respuesta_central.append(tiempo_respuesta_central)

tiempo_medio_respuesta_central = np.mean(tiempos_respuesta_central)

# Crear gráfica de barras
plt.bar(range(num_simulaciones), tiempos_respuesta_central)
plt.axhline(y=tiempo_medio_respuesta_central, color='r', linestyle='--', label='Tiempo Medio')

# Configurar etiquetas y título del gráfico
plt.xlabel('Simulación')
plt.ylabel('Tiempo de Respuesta')
plt.title('Tiempo Medio de Respuesta de la Central de Emergencias')

# Mostrar leyenda
plt.legend()

# Mostrar gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal para cada operador
medias = [2.0, 2.1, 1.9]
desviaciones = [0.4, 0.3, 0.5]

num_simulaciones = 2000
num_llamadas = 100

tiempos_respuesta_central = []
tiempos_medios_convergencia = []

for _ in range(num_simulaciones):
    tiempos_respuesta_operadores = []

    for _ in range(num_llamadas):
        tiempos_llamadas = np.random.normal(medias, desviaciones)
        tiempo_respuesta_operador = np.min(tiempos_llamadas)
        tiempos_respuesta_operadores.append(tiempo_respuesta_operador)

    tiempo_respuesta_central = np.mean(tiempos_respuesta_operadores)
    tiempos_respuesta_central.append(tiempo_respuesta_central)

    tiempo_medio_convergencia = np.mean(tiempos_respuesta_central)
    tiempos_medios_convergencia.append(tiempo_medio_convergencia)

tiempo_medio_respuesta_central = np.mean(tiempos_respuesta_central)

# Crear gráficas
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Gráfica de barras
ax1.bar(range(num_simulaciones), tiempos_respuesta_central)
ax1.axhline(y=tiempo_medio_respuesta_central, color='r', linestyle='--', label='Tiempo Medio')
ax1.set_ylabel('Tiempo de Respuesta')
ax1.set_title('Tiempo Medio de Respuesta de la Central de Emergencias')
ax1.legend()

# Gráfica de convergencia
ax2.plot(range(1, num_simulaciones + 1), tiempos_medios_convergencia)
ax2.set_xlabel('Número de Simulaciones')
ax2.set_ylabel('Tiempo Medio de Respuesta')
ax2.set_title('Convergencia del Tiempo Medio de Respuesta')

# Ajustar espaciado entre subgráficas
plt.tight_layout()

# Mostrar gráficas
plt.show()
```
