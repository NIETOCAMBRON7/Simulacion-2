Simulación mediante eventos discretos

Para realizarla, damos seguimiento a ciertas variables.

1: Variable de tiempo t: Se refiere al tiempo (simulado) que ha transcurrido

2: Variables de conteo: Numero de veces que han ocurrido ciertos eventos hasta el momento t

3: Variable de estado del sistema: Describe el estado del sistema en el tiempo t.

```python
import random
import math
import matplotlib.pyplot as plt

def generar_numeros_aleatorios_uniformes(n):
    numeros = []
    for _ in range(n):
        numeros.append(random.uniform(0, 1))
    return numeros

def calcular_resultados(numeros, lambda_value):
    resultados = []
    for num in numeros:
        resultado = (-1 / lambda_value) * math.log(num)
        resultados.append(resultado)
    return resultados

# Parámetros
n = 1000000  # Cantidad de números aleatorios a generar
lambda_value = 2.0  # Valor de lambda

# Generar números aleatorios
numeros_aleatorios = generar_numeros_aleatorios_uniformes(n)


# Graficar histograma de mis numeros aleatorios
plt.hist(numeros_aleatorios, bins=20, alpha=0.5)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Números Aleatorios Uniformes')
plt.show()



# Calcular resultados
resultados = calcular_resultados(numeros_aleatorios, lambda_value)

# Graficar histograma
plt.hist(resultados, bins='auto', edgecolor='orange')
plt.xlabel('(-1/λ)log(U)')
plt.ylabel('Frecuencia')
plt.title('Histograma de (-1/λ)log(U)')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def generar_numeros_exponenciales(n, lambd):
    numeros = []
    for _ in range(n):
        numeros.append(random.expovariate(lambd))
    return numeros

# Parámetros
n = 1000000  # Cantidad de números a generar
lambd = 0.5  # Tasa de la distribución exponencial (1/lambda con lambda = 2)

# Generar números exponenciales
numeros_exponenciales = generar_numeros_exponenciales(n, lambd)

# Graficar histograma
plt.hist(numeros_exponenciales, bins=100, alpha=0.5)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Números Exponenciales')
plt.show()
```

```python
import math
import random
import matplotlib.pyplot as plt
def generar_poisson(T, lambd):
    t = 0
    I = 0
    S = []

    while t <= T:
        U = random.uniform(0, 1)
        t += -math.log(U) / lambd

        if t > T:
            break

        I += 1
        S.append(t)

    return S

# Ejemplo de uso
T = 10  # Unidades de tiempo
lambd = 1.5  # Tasa promedio de eventos por unidad de tiempo

proceso_poisson = generar_poisson(T, lambd)
print("Proceso de Poisson generado:", proceso_poisson)
```

```python
import math
import random

def generar_poisson_process(T, lambd):
    t = 0
    I = 0
    S = []

    while t < T:
        U = random.uniform(0, 1)
        t_n = t - (1 / lambd) * math.log(U)

        if t_n > T:
            break

        t = t_n
        I += 1
        S.append(t)

    return S

# Ejemplo de uso
T = 10  # Unidades de tiempo
lambd = 10  # Tasa promedio de eventos por unidad de tiempo

proceso_poisson = generar_poisson_process(T, lambd)

print("Proceso de Poisson generado:")
for i, t in enumerate(proceso_poisson):
    print("Evento {}: tiempo {}".format(i+1, t))
```

El tiempo entre llegadas de piezas a una estación de ensamblado sigue una distribu-
ción exponencial con media 2 minutos. La duración de la tarea llevada a cabo por el operario en dicha estación adopta una distribución exponencial con media 1 minuto
por pieza. Determine el tiempo medio de permanencia de las piezas en dicho proceso.

```python
import random

def tiempo_llegada_piezas():
    return random.expovariate(0.5)  # Media de 2 minutos

def tiempo_tarea():
    return random.expovariate(1)  # Media de 1 minuto

def tiempo_permanencia(num_piezas):
    tiempo_total = 0
    for _ in range(num_piezas):
        tiempo_llegada = tiempo_llegada_piezas()
        tiempo_tarea_realizacion = tiempo_tarea()
        tiempo_total += tiempo_llegada + tiempo_tarea_realizacion
    return tiempo_total / num_piezas

num_piezas = 1000  # Número de piezas para simular

tiempo_medio_permanencia = tiempo_permanencia(num_piezas)
print("Tiempo medio de permanencia de las piezas en el proceso:", tiempo_medio_permanencia, "minutos")
```
