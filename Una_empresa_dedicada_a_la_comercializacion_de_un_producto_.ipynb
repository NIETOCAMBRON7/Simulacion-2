Una empresa dedicada a la comercializacion de un producto de mucho exito quiere determinar el numero de unidades que debe comprar con el objetivo de satisfacer la demanda de sus clientes los proximos meses

la distribucion de las ventas de los ultimos 3 años se recoge en la tabla siguiente:

| Unidades vendidas | 100 | 200 | 300 |  400 | 500 | 600 |
|-------------------|-----|-----|-----|------|-----|-----|
|Frecuencia en meses|  8  |  9  |  4  |   9  |   4 |   2 |

el coste de cada unidad de productos asciende a 750 euros, siendo su precio de venta unitario de 1000 euros. Se sabe que dentro de un mes saldra al mercado un nuevo modelo de caracteristicas muy superiores al modelo actual.
Cuando eso ocurra la empresa podra devolver al distribuidor las unidades sobrantes del modelo actual y este le indemnizara con 250 euros por cada unidad retornada

### Primero.-

Vamos a generar un numero aleatorio entre 0 y 1 para simular el numero de ventas que se puedan llegar a generar para el proximo mes


```python
import random

def generar_numero_aleatorio():
    return random.random()
```

```python
def obtener_unidades_vendidas(numero_aleatorio):
    if numero_aleatorio <= 0.22:
        return 100
    elif numero_aleatorio <= 0.47:
        return 200
    elif numero_aleatorio <= 0.58:
        return 300
    elif numero_aleatorio <= 0.83:
        return 400
    elif numero_aleatorio <= 0.94:
        return 500
    else:
        return 600
```

```python
def beneficio(pedido):
    numero = generar_numero_aleatorio()

    # Obtener el valor de "Unidades Vendidas"
    unidades_vendidas = obtener_unidades_vendidas(numero)


    #print("Número aleatorio:", numero)
    #print("Unidades vendidas:", unidades_vendidas)

    if unidades_vendidas > pedido:
      U_devueltas = 0
    else:
      U_devueltas = pedido - unidades_vendidas

    Beneficio = unidades_vendidas*1000 - pedido*750 + U_devueltas*250
    #print ("Beneficio:" ,Beneficio)
    return Beneficio
```

```python
beneficio(400)
```

```python
def simulacion(num_simulaciones,n):
    L = []
    for i in range (num_simulaciones):
      L.append(beneficio(n))
    return sum(L)/len(L),L
```

```python
def simulacion(num_simulaciones,n):
    L = []
    benef = 0
    for i in range (num_simulaciones):
      benef += beneficio(n)
      L.append(benef/(i+1))
    return L
```
