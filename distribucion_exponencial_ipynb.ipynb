```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 15 números aleatorios
random_numbers = np.random.rand(15)

# Generar números con distribución exponencial
exponential_numbers = np.random.exponential(scale=1.0, size=1000)

# Graficar los números generados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.hist(random_numbers, bins=15, color='blue', alpha=0.7)
plt.title('Números Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(exponential_numbers, bins=30, color='green', alpha=0.7)
plt.title('Distribución Exponencial')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Este código utiliza la función np.random.rand() para generar los 15 números aleatorios y la función np.random.exponential() para generar los números que siguen una distribución exponencial. Luego, crea un gráfico con dos subplots: uno para mostrar la distribución de los números aleatorios y otro para mostrar la distribución exponencial.

ahora, notemos que si hacemos mas cantidad de numeros aleatorios, el histograma de numeros aleatorios es mas parejo, y el de la distribucion exponencial se asemeja a una curva exponencial.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 15 números aleatorios
random_numbers = np.random.rand(1000000)  # un millon de datos

# Generar números con distribución exponencial
exponential_numbers = np.random.exponential(scale=1.0, size=1000)

# Graficar los números generados
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.hist(random_numbers, bins=15, color='blue', alpha=0.7)
plt.title('Números Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(exponential_numbers, bins=30, color='green', alpha=0.7)
plt.title('Distribución Exponencial')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
